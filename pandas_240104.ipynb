```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# total_bill의 중간값 이하인 것만 조회
print(tips['total_bill'].median())
tips.loc[tips['total_bill'] <= tips['total_bill'].median(), :]
```

```python
# day가 Sun인 데이터만 조회하기
tips.loc[tips['day'] == 'Sun', :]
```

```python
# tip의 평균보다 크면서 , time이 Dinner인 데이터만 조회
tips.loc[(tips['tip'] > tips['tip'].mean()) & (tips['time'] == 'Dinner'), :]
```

```python
print(tips['tip'].mean())
tips.loc[(tips['tip']>tips['tip'].mean())|
        (tips['time']=='Dinner')&
        (tips['size']==3)&
        (tips['sex']=='Male')&
        (tips['day']=='Sat'),
:]
```
